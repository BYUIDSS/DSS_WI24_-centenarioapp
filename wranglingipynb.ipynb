```python
import pandas as pd

```

```python
centenario = pd.read_csv("raw_data/centenario2019.csv")
centenario1 = pd.read_csv("raw_data/centenario2020.csv")
centenario2 = pd.read_csv("raw_data/centenario2021.csv")
centenario3 = pd.read_csv("raw_data/centenario2022.csv")
centenario4 = pd.read_csv("raw_data/centenario2023.csv")

centenario_final = pd.concat([centenario, centenario1, centenario2, centenario3, centenario4], ignore_index=True)

```

```python
prices = pd.read_csv("raw_data/trips_price2019.csv")
prices1 = pd.read_csv("raw_data/trips_price2020.csv")
prices2 = pd.read_csv("raw_data/trips_price2021.csv")
prices3 = pd.read_csv("raw_data/trips_price2022.csv")
prices4 = pd.read_csv("raw_data/trips_price2023.csv")

prices_final = pd.concat([prices, prices1, prices2, prices3, prices4], ignore_index=True)

```

```python
def categorize_car_type(car_name):
    car_types = {
        "Compact": ["Jetta", "KIA RIO", "Prius", "Vento"],
        "Large": ["Passat", "Camry"],
        "SUV": ["Suburban"],
        "Passanger": ["Transit", "Econoline"]
    }

    # Check each car type for a match in the car name
    for car_type, names in car_types.items():
        for name in names:
            if name.lower() in car_name.lower():  # Case insensitive match
                return car_type
    return "Unknown"  # If no matches were found

centenario_final['Type'] = centenario_final['Car'].apply(categorize_car_type)

centenario_final['Year'] = pd.to_datetime(centenario_final['Date']).dt.year



centenario_final.head()

centenario_final.to_csv("raw_data/centanario_trips.csv")
```

```python
# Melt the cleaned dataframe again
long_prices = pd.melt(prices_final, id_vars=["type", "year"], var_name="destination", value_name="price")

# Display the first few rows of the correctly reshaped dataframe
long_prices.to_csv("raw_data/prices_file.csv")
```
